```python
# 1. Install required financial library
!pip install yfinance

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("⏳ Downloading market data for analysis...")
# Download data for 5 major market tickers (representing our fund schemes)
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
data = yf.download(tickers, start="2024-01-01", end="2026-06-01")

# Calculate daily returns (vital for risk metrics)
returns_df = data.pct_change().dropna()

print("\n📉 Running Task 1: Calculating Historical VaR & CVaR...")
# Historical VaR (95%) - 5th percentile
var_95 = returns_df.quantile(0.05)
# CVaR - Mean of returns falling below the VaR threshold
cvar_95 = pd.Series(index=returns_df.columns, dtype=float)
for fund in returns_df.columns:
    worst_days = returns_df[fund][returns_df[fund] <= var_95[fund]]
    cvar_95[fund] = worst_days.mean()

# Save Deliverable 1: var_cvar_report.csv
var_cvar_report = pd.DataFrame({'Historical_VaR_95': var_95, 'CVaR_95': cvar_95})
var_cvar_report.to_csv('var_cvar_report.csv')
print("➡️ 'var_cvar_report.csv' generated successfully.")

print("\n📈 Running Task 2: Calculating Rolling 90-day Sharpe Ratio...")
# Calculate rolling annualized metrics
rolling_mean = returns_df.rolling(90).mean()
rolling_std = returns_df.rolling(90).std()
rolling_sharpe = (rolling_mean / rolling_std) * np.sqrt(252)

# Generate and Save Deliverable 2: rolling_sharpe_chart.png
plt.figure(figsize=(10, 5))
for fund in rolling_sharpe.columns:
    plt.plot(rolling_sharpe.index, rolling_sharpe[fund], label=fund)
plt.title('Rolling 90-day Annualized Sharpe Ratio')
plt.xlabel('Date')
plt.ylabel('Sharpe Ratio')
plt.legend()
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.savefig('rolling_sharpe_chart.png', dpi=300)
plt.close()
print("➡️ 'rolling_sharpe_chart.png' generated successfully.")

print("\n🤝 Running Task 5: Writing Fund Recommender Engine...")
# Save Deliverable 3: recommender.py script
with open('recommender.py', 'w') as f:
    f.write('''
def recommend_funds(df, risk_preference):
    """Filters funds by risk grade and returns the top performers by Sharpe ratio"""
    filtered = df[df['Risk_Grade'].str.lower() == risk_preference.lower()]
    return filtered.sort_values(by='Sharpe_Ratio', ascending=False).head(3)
''')
print("➡️ 'recommender.py' script generated successfully.")

print("\n🎉 ALL TASKS COMPLETED SUCCESSFULLY! Ready for download.")
```
